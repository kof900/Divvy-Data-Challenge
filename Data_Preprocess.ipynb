# Pull Weather Information using Dark Sky API

Due to the limited amount of available features, I pulled the daily and hourly weather information from 2016 Q3 to 2017 Q2. I used the Dark Sky API. It allows 1000 free request per day, which is enough to cover from 2016 Q3 to 2017 Q2. Note that, I didn't treat daily information as "pre-known" information since we could use the forcast as its data. It would not bring in the time machine.

```python
import json, requests, sys
import datetime
import pytz
import calendar
```

```python
# 1467349200 is the unix time for 2016/7/01 00:00:00 CST
import time
time_0 = 1467349200
time_0
```

```python
# Make a get request to get the latest position of the international space station from the opennotify api
# chicago latitude longitude 41.8781,-87.6298

# dark sky API key
key0 = '7b9dd22e760f21ef06a5e59ccb97fa22'
key1 = 'b22052ca483684ed023b91654c57e100'

def unix_secs(y, m, d, h):
    """given year, month, day, hour from Central timezone, return unix timestamp"""
    return calendar.timegm(datetime.datetime(y, m, d, h, tzinfo=pytz.timezone('America/Chicago')).utctimetuple())

def weatherDaily(time, key):
    response = requests.get("https://api.darksky.net/forecast/%s/41.8781,-87.6298,%d?exclude=currently, minutely, alerts, flags" % (key, time))
    
    # print the status code of the response
    print(response.status_code)
    response.content.decode("utf-8")
    
    # loads the pulled dataset as json format
    weatherData = json.loads(response.text)
    df_daily = pd.DataFrame([list(weatherData['daily']['data'][0].values())] * 24,
                            columns=list(map(lambda x: 'daily_' + x, list(weatherData['daily']['data'][0].keys()))))
    
    df_hour = pd.DataFrame()
    for i in range(0, 24):
        tmp = pd.Series(weatherData['hourly']['data'][i])
        df_hour = pd.concat([df_hour, tmp.to_frame().T], ignore_index=True)
    
    df_out = pd.concat([df_hour, df_daily], axis=1)
    df_out['unix_secs'] = [time + 3600 * x for x in range(24)]
    
    return(df_out)
```

```python
%%time
df_weather = pd.DataFrame()
for day in range(0, 365):
    print(day)
    tmp = weatherDaily(day * 86400 + time_0, key0)
    df_weather = pd.concat([df_weather, tmp], ignore_index=True)
```
